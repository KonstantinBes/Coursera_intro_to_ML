```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
```

```python
data = pd.read_csv('D:/Work/Data_files/working_dir/wine.data', header=None)
data.head()
```

```python
X = data.drop(data.columns[0], axis = 1)
X.head()
```

```python
y = data[data.columns[0]]
y.head()
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
stats = {}
for k in range(1, 51):
    clf = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(clf, X, y, cv=kf.split(X, y))    # del y?
    score = np.append(score, score.mean())
    stats[k] = score
df = pd.DataFrame.from_dict(stats, orient='index').reset_index()\
    .rename(columns = {df.columns[0]: 'k',
                      df.columns[1]: 'cv_1',
                      df.columns[2]: 'cv_2',
                      df.columns[3]: 'cv_3',
                      df.columns[4]: 'cv_4',
                      df.columns[5]: 'cv_5',
                      df.columns[6]: 'mean_accuracy'})
df.head()
```

```python
df[df['mean_accuracy'] == df['mean_accuracy'].max()]
```

```python
X_normalized = pd.DataFrame(scale(X), columns = X.columns)
X_normalized
```

```python
stats_normalized = {}
for k in range(1, 51):
    clf = KNeighborsClassifier(n_neighbors=k)
    score = cross_val_score(clf, X_normalized, y, cv=kf.split(X_normalized, y))
    score = np.append(score, score.mean())
    stats_normalized[k] = score
df_norm = pd.DataFrame.from_dict(stats_normalized, orient='index').reset_index()
df_norm.head()
```

```python
df_norm = df_norm.rename(columns = {df_norm.columns[0]: 'k',
                                   df_norm.columns[1]: 'cv_1',
                                   df_norm.columns[2]: 'cv_2',
                                   df_norm.columns[3]: 'cv_3',
                                   df_norm.columns[4]: 'cv_4',
                                   df_norm.columns[5]: 'cv_5',
                                   df_norm.columns[6]: 'mean_accuracy'})
df_norm.head()
```

```python
df_norm.sort_values('mean_accuracy', ascending = False)
```
