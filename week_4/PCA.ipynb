```python
import pandas as pd
import numpy as np

from sklearn.decomposition import PCA
from numpy import corrcoef
from numpy import concatenate
```

```python
close_prices = pd.read_csv('D:/Work/Data_files/working_dir/close_prices.csv')
index_col = list(close_prices['date'].values)
close_prices.set_index(keys='date', inplace=True)
close_prices.head()
```

```python
pca = PCA(n_components=10)
pca.fit(close_prices)
```

```python
pca.explained_variance_ratio_
```

```python
close_prices_transformed = pd.DataFrame(pca.transform(close_prices), index=index_col)
close_prices_transformed
```

```python
dj_index = pd.read_csv('D:/Work/Data_files/working_dir/djia_index.csv')\
    .set_index('date')
dj_index.head()
```

```python
price_index_df = pd.concat([close_prices_transformed[0], dj_index], axis=1)
main_component = list(price_index_df[0].to_numpy())
main_component[0:10]
```

```python
dow_jones_index = list(price_index_df['^DJI'].to_numpy())
dow_jones_index[0:10]
```

```python
correlate = np.array([main_component, dow_jones_index])
correlate[0][0:10]
```

```python
corrcoef(main_component, dow_jones_index)
```
