```python
# https://www.coursera.org/learn/vvedenie-mashinnoe-obuchenie/programming/u08ys/priedobrabotka-dannykh-v-pandas

import pandas as pd


data = pd.read_csv('D:/Work/Data_files/working_dir/titanic.csv', index_col = 'PassengerId')
data.head()

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html
# https://www.kaggle.com/c/titanic/data
# https://bitbucket.org/hrojas/learn-pandas
```

```python
# 1. Quantity of man and women on board (two numbers space separated):

men_num = data.groupby('Sex').count()
#women_num = data['Sex'].count('female')

print(men_num)
```

```python
# 2. Share of survived (number without % token, two decimal accuracy):
survived = list(data['Survived']).count(1)
died = list(data['Survived']).count(0)
surv_rate = survived/(survived + died) * 100
surv_rate
```

```python
# 3. Share of first-class passengers (number without % token, two decimal accuracy):

first_class = data['Pclass'][data.Pclass == 1].count()
total = data['Pclass'].count()
share_first_class = (first_class / total) * 100
share_first_class
```

```python
# 4. Mean and median of passenger`s age (two numbers space separated):
 
mean_age = data['Age'].mean()
print("Mean_age: %f." %mean_age)
median_age = data['Age'].median()
print("Median_age: %f." %median_age)
data['Age'].describe()
```

```python
# 5. Pearson correlation between SibSp and Parch (number, two decimal accuracy):

data[['SibSp', 'Parch']].corr(method = 'pearson') #select_dtypes(exclude = 'object')
```

```python
# 6. Most popular women First Name on board (string):

import collections


women_names = list(data.query("Sex == 'female'")['Name'])
first_names = collections.Counter()
first_name = ''
df = {'Full': [], 'First': []}
prefixes = ['Miss. ', '(']

for name in women_names:
    for prefix in prefixes:
        if(prefix in name):
            start_index = name.find(prefix) + len(prefix)
            # Lady has one given name case:
            if not name[start_index::].find(' '):
                first_name = name[start_index:]
            # Lady has more than one given names case:
            else:
                stop_index = start_index + name[start_index::].find(' ')
                first_name = name[start_index: stop_index]
            break
        else:
             first_name = ''
    if(first_name == ''):
        first_name = name.split(' ')[0]
    df['Full'].append(name)
    df['First'].append(first_name)
    first_names[first_name] += 1

#pd.DataFrame(df)
most_popular_name = first_names.most_common()
most_popular_name
```
